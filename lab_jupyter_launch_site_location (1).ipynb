```python
# TASK 3: Calculate the distances between a launch site to its proximities
import pandas as pd
import numpy as np

# Supongamos que ya tienes un DataFrame de sitios de lanzamiento
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates()

# Crea un nuevo DataFrame para almacenar las distancias
distances_list = []  # Lista para almacenar las filas

# Extrae las coordenadas de los sitios de lanzamiento
latitudes = launch_sites['Lat'].values
longitudes = launch_sites['Long'].values

# Crea un DataFrame de pares de sitios
for i in range(len(launch_sites)):
    for j in range(len(launch_sites)):
        if i != j:  # Evita calcular la distancia consigo mismo
            # Extrae las coordenadas del sitio actual y del sitio cercano
            lat1 = latitudes[i]
            lon1 = longitudes[i]
            lat2 = latitudes[j]
            lon2 = longitudes[j]

            # Calcula la distancia usando la fórmula de Haversine
            dlat = np.radians(lat2 - lat1)
            dlon = np.radians(lon2 - lon1)

            a = np.sin(dlat / 2)**2 + np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * np.sin(dlon / 2)**2
            c = 2 * np.arcsin(np.sqrt(a))

            # Radio de la Tierra en kilómetros
            r = 6371  
            distance = c * r
            
            # Añade la distancia a la lista como un diccionario
            distances_list.append({
                'Launch Site': launch_sites.iloc[i]['Launch Site'],
                'Nearby Site': launch_sites.iloc[j]['Launch Site'],
                'Distance (km)': distance
            })

# Convierte la lista de diccionarios a un DataFrame
distances_df = pd.DataFrame(distances_list)

# Muestra el DataFrame de distancias
print(distances_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
